```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

def one_hot_encode_columns(df, columns_to_encode):
    """
    One-hot encodes specified columns in a pandas DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        columns_to_encode (list): A list of column names to one-hot encode.

    Returns:
        pd.DataFrame: The DataFrame with one-hot encoded columns.
    """

    encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse_output=False for a numpy array, handle_unknown ignores unknown categories.
    encoded_data = encoder.fit_transform(df[columns_to_encode])
    encoded_columns = encoder.get_feature_names_out(columns_to_encode) # get the new column names.
    encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

    # Concatenate the encoded columns with the original DataFrame
    df = pd.concat([df.drop(columns_to_encode, axis=1).reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)

    return df
```

```python
# Specify the columns to one-hot encode
columns_to_encode = ["family_status","income_type", "education"]

# One-hot encode the columns
encoded_df = one_hot_encode_columns(df, columns_to_encode)

# Print the encoded DataFrame
print(encoded_df)
```

```python
import pandas as pd
from autogluon.tabular import TabularPredictor

from sklearn.preprocessing import StandardScaler

def preprocess_stock_data(train_data):
    # Specify the target column
    target = 'primary_income'

    # Initialize the TabularPredictor with time_limit
    predictor = TabularPredictor(label=target, eval_metric='mean_absolute_percentage_error').fit(train_data, time_limit=300)

    return predictor
```

```python
# Predict stock prices
predictor = preprocess_stock_data(encoded_df)
```

```python
test_data = pd.read_csv("/content/test.csv")
```

```python
# prompt: preprocess like the train data to test_data

def preprocess_test_data(test_data, predictor):
    # Perform the same preprocessing steps as the training data
    test_data["family_status"].fillna("Single", inplace=True)
    test_data["max_outstanding_balance_12_months"].fillna(0, inplace=True)

    columns_to_encode = ["family_status", "income_type", "education"]
    test_encoded_df = one_hot_encode_columns(test_data[columns_to_encode + ["max_outstanding_balance_12_months"]], columns_to_encode)

    # Ensure the test data has the same columns as the training data after one-hot encoding
    missing_cols = set(encoded_df.columns) - set(test_encoded_df.columns)
    for c in missing_cols:
        test_encoded_df[c] = 0
    test_encoded_df = test_encoded_df[encoded_df.columns]

    # Scale numerical features using the scaler fit on the training data
    test_encoded_df["max_outstanding_balance_12_months"] = scaler.transform(test_encoded_df[["max_outstanding_balance_12_months"]])

    return test_encoded_df

# Preprocess the test data
test_preprocessed = preprocess_test_data(test_data.copy(), predictor)

# Now you can use test_preprocessed for predictions
# Example: predictions = predictor.predict(test_preprocessed)
```

```python
test_preprocessed.drop("primary_income", axis=1, inplace=True)
```

```python
from autogluon.tabular import TabularPredictor
from tqdm import tqdm  # For displaying progress bar

def predict_in_batches(predictor, test_data, batch_size=1000):
    """
    Predicts in batches and displays a progress bar.

    Args:
        predictor (TabularPredictor): The trained predictor.
        test_data (pd.DataFrame): The data to predict on.
        batch_size (int): The size of each batch.

    Returns:
        pd.Series: The predictions.
    """
    num_batches = (len(test_data) + batch_size - 1) // batch_size
    predictions = []

    # Use tqdm to display a progress bar
    with tqdm(total=len(test_data), desc="Predicting") as pbar:
        for i in range(num_batches):
            start_idx = i * batch_size
            end_idx = min((i + 1) * batch_size, len(test_data))
            batch_predictions = predictor.predict(test_data.iloc[start_idx:end_idx])
            predictions.extend(batch_predictions)
            pbar.update(end_idx - start_idx)  # Update progress bar

    return pd.Series(predictions, index=test_data.index)

# Example usage:
predictions = predict_in_batches(predictor, test_preprocessed)
```
